# Introduction to Pandas

### What is Pandas?
Pandas is a library that allows you to run data analysis and manipulation in preparation for application to machine learning models (one example)

### Why Pandas?
1. Simple to use
2. Integrated with many other data science & ML Python tools
3. Helps you get your data ready for machine learning - raw data is often illfit for ML applications

### Filling Missing Data

```python
car_sales_missing = pd.read_csv("data/car-sales-missing-data.csv")
car_sales_missing
```

```python
car_sales_missing.dropna(inplace=True)
car_sales_missing
```

```python
# Column from series
seats_column = pd.Series([5, 5, 5, 5, 5])
seats_column.dtype
```

```python
# New column
car_sales["Seats"] = seats_column
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
```

```python
car_sales
```

```python
# Column from Python list
fuel_economy = [7.5, 9.2, 5.0, 9.6, 8.7, 4.7, 8.7, 3.0, 4.5, 3.4]
car_sales["Fuel per"] = fuel_economy
car_sales
```

```python
car_sales["Total fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per"]
car_sales
```

```python
# Create a column from a single value
car_sales["Number of wheels"] = 4
car_sales
```

```python
car_sales["Passed safety check"] = True
car_sales
```

```python
car_sales = car_sales.drop("Total fuel used", axis=1)
car_sales
```

```python
car_sales["Has Bluetooth"] = True
car_sales
```

```python
car_sales_shuffled = car_sales.sample(frac=1) # shuffles 100% of the data order
car_sales_shuffled
```

```python
car_sales
```

```python
car_sales_shuffled.reset_index(inplace=True, drop=True)
car_sales_shuffled
```

```python
# lambda is a keyword in Python for an anonymous function
# .apply is a way to apply a function to a column
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x / 1.6)
car_sales
```
